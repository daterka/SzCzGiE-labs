### Part V

Chaos game (Sierpinski's Triangle)

```python
import matplotlib.pyplot as plt
from random import random, randint
import numpy as np
import matplotlib.cm as cm

sqrt = lambda x: x ** 0.5
zeros = lambda N: [0 for x in range(N)]


def midpoint(p, q):
    return (0.5*(p[0] + q[0]), 0.5*(p[1] + q[1]))

# Three corners of an equilateral triangle
corner = [(0, 0), (0.5, sqrt(3)/2), (1, 0)]

N = 1000
x = zeros(N)
y = zeros(N)

x[0] = random()
y[0] = random()
for i in range(1, N):
    k = randint(0, 2) # random triangle vertex
    x[i], y[i] = midpoint( corner[k], (x[i-1], y[i-1]) )

colors = cm.rainbow(np.linspace(0, 1, len(y)))

plt.figure(figsize=(20, 8))
plt.title(label='Sierpinski\'s Triangle')
plt.xlabel('x')
plt.ylabel('y')

for x, y, c in zip(x, y, colors):
    plt.scatter(x, y, color=c)
    
# plt.scatter(x, y)
# plt.figure(figsize=(20, 8))
plt.show()
```

### Part V

Chaos in linear equations

```python
import matplotlib.pyplot as plt
from random import random, randint
import numpy as np
import matplotlib.cm as cm


a = 3.7
N = 1000

sqrt = lambda x: x ** 0.5
zeros = lambda N: [0 for x in range(N)]
foo = lambda x: a*x*(1-x)


x1 = zeros(N)
y1 = zeros(N)
x1[0] = 0.22222228
y1[0] = foo(x1[0])

x2 = zeros(N)
y2 = zeros(N)
x2[0] = 0.22222227
y2[0] = foo(x2[0])

for i in range(1, N):   
    x1[i] = y1[i-1]
    y1[i] = foo(x1[i-1])
    x2[i] = y2[i-1]
    y2[i] = foo(x2[i-1])

# colors = cm.rainbow(np.linspace(0, 1, len(y)))

# for x, y, c in zip(x, y, colors):
#     plt.scatter(x, y, color=c)

x = [x for x in range(N)]
```

```python
# scatter plot of foo trajectories for two different a parameters
plt.figure(figsize=(20, 8))
plt.title(label='scatterplot of f(x)=x*(1-x)')
plt.xlabel('iteration')
plt.ylabel('f(x)')

plt.scatter(x, y1, color='cyan', alpha=0.35, linewidths=0.5, edgecolors='blue', label=0.22222228)
plt.scatter(x, y2, color='pink', alpha=0.35, linewidths=0.5, edgecolors='purple', label=0.22222227)

plt.legend()
plt.show()
```

Jak wida na wykresie punktowym minimalna zmiana argumentu startowego systemu sprawia, e cho pocztkowo nakadajce si na siebie punkty w przypadku obu uruchomie wraz ze wzrostem iteracji zaczy si od siebie oddala, osigajc w kocowych iteracjach zupenie rne wsprzdne na paszczynie.

```python
# line2d plot of foo trajectories for two different a parameters
plt.figure(figsize=(20, 8))
plt.title(label='line2d plot of f(x)=x*(1-x)')
plt.xlabel('iteration')
plt.ylabel('f(x)')

plt.plot(x, y1, color='cyan', alpha=0.6, label=0.22222228)
plt.plot(x, y2, color='pink', alpha=0.9, label=0.22222227)

plt.legend()
plt.show()
```

Powyszy wikres obrazuje krzyw zmian wartoci obydwu ukadw, rnicych si wartoci stanu pocztkowego 8 cyfr rozwinicia dziesitnego.

```python
# zoomed (x4)line2d plot of foo trajectories for two different a parameters
plt.figure(figsize=(20, 8))
plt.title(label='zoomed (x4) line2d plot of f(x)=x*(1-x)')
plt.xlabel('iteration')
plt.ylabel('f(x)')

plt.xlim(right=N/4)

plt.plot(x, y1, color='cyan', alpha=0.6, label=0.22222228)
plt.plot(x, y2, color='pink', alpha=0.9, label=0.22222227)

plt.legend()
plt.show()
```
