```python
def logRateOfReturn(data=[]):
    logReturnRate=[]
    for i in range(1, len(data)):
        logReturnRate.append(log(data[i]/data[i-1]))

    return pd.DataFrame({'logReturnRate': np.array([0.0] + logReturnRate)})
```

```python
def z_score(df):
    df_std = df.copy()  
    return (df_std - df_std.mean()) / df_std.std()
```

```python
def generateWhiteNoise(mean=0, std=1, size=1000):
    return pd.DataFrame(np.random.normal(mean, std, size=size))
```
